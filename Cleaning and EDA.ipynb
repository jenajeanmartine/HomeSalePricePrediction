### Null Values & Vectorizing Qualitative/Ordinal Strings

#### Found Columns in Data Dictionary where NA means the absence of a feature and change to 0

```python
cols_NA_notnull = ['alley', 'fireplace_qu', 'pool_qc', 'fence', 'misc_feature', 'bsmt_cond', 'garage_qual', 'garage_cond']
df.loc[:, cols_NA_notnull] = df.loc[:, cols_NA_notnull].fillna(0)
#df.isnull().sum().sort_values(ascending = False).iloc[0:19]
```

#### Investigate column values for vectorizing strategy

```python
# for col in cols_NA_notnull:
#     print(col,': ',df.groupby(col)[col].count(), """
#     """)
```

#### Programmatically vectorize cols with the same 5 quality categories

```python
quals = ['garage_cond', 'garage_qual', 'bsmt_cond', 'fireplace_qu', 'exter_qual', 'exter_cond', 'kitchen_qual']
```

```python
for col in quals:
    df.loc[df[col] == 'Ex', col] = 5
    df.loc[df[col] == 'Gd', col] = 4
    df.loc[df[col] == 'TA', col] = 3
    df.loc[df[col] == 'Fa', col] = 2
    df.loc[df[col] == 'Po', col] = 1
    df.loc[df[col] == 'NA', col] = 0
```

#### Deal with the remaining four columns:

#### PoolQC - change ordinal categories to numerical.  
       Ex	Excellent = 4, 
       Gd	Good = 3, 
       TA	Average/Typical = 2, 
       Fa	Fair = 1, 
       NA	No Pool = 0

```python
df.loc[df['pool_qc'] == 'Ex', 'pool_qc'] = 4
df.loc[df['pool_qc'] == 'Gd', 'pool_qc'] = 3
df.loc[df['pool_qc'] == 'TA', 'pool_qc'] = 2
df.loc[df['pool_qc'] == 'Fa', 'pool_qc'] = 1
df.loc[df['pool_qc'] == 'NA', 'pool_qc'] = 0
```

#### Alley: <br/>
NA = none = 0 <br/>
Grvl = gravel = 1 <br/>
Pave = paved = 2 <br/>


```python
df.loc[df['alley'] == 'Pave', 'alley'] = 2
df.loc[df['alley'] == 'Grvl', 'alley'] = 1
df.loc[df['alley'] == 'NA', 'alley'] = 0
```

#### Fences - vectorize <br/>
good privacy = GdPrv = 2<br/>
good wood = GdWo = 2<br/>
minimum privacy = MnPrv = 1<br/>
minimum wood/wire = MnWw = 1<br/>
no fence = NA = 0<br/>
Decided to group into 2 qualities of fence or no fence. Alternately could do 0,1,2,3 with Goodwood as second best to good privacy

```python
df.loc[(df['fence'] == 'GdPrv') | (df['fence'] == 'GdWo'), 'fence'] = 2
df.loc[(df['fence'] == 'MnPrv') | (df['fence'] == 'MnWw'), 'fence'] = 1
df.loc[df['fence'] == 'NA', 'fence'] = 0
```

### Vectorize Categorical Garage Columns 

#### Garage Finish: <br/>
Fin	Finished = 3<br/>RFn	Rough Finished = 2<br/>	Unf	Unfinished = 1<br/>NA	No Garage = 0<br/>

```python
df.loc[df['garage_finish'] == 'Fin', 'garage_finish'] = 3
df.loc[df['garage_finish'] == 'RFn', 'garage_finish'] = 2
df.loc[df['garage_finish'] == 'Unf', 'garage_finish'] = 1
df.loc[df['garage_finish'].isnull(), 'garage_finish'] = 0
```

#### Functional to numeric

```python
df.loc[df['functional'] == 'Typ', 'functional'] = 7
df.loc[df['functional'] == 'Min1', 'functional'] = 6
df.loc[df['functional'] == 'Min2', 'functional'] = 5
df.loc[df['functional'] == 'Mod', 'functional'] = 4
df.loc[df['functional'] == 'Maj1', 'functional'] = 3
df.loc[df['functional'] == 'Maj2', 'functional'] = 2
df.loc[df['functional'] == 'Sev', 'functional'] = 1
df.loc[df['functional'] == 'Sal', 'functional'] = 0
```

### Create log columns of quantifiable and non-zero cols

```python
df['log_saleprice'] = np.log(df['saleprice'])
```

```python
to_log = ['lot_area','1st_flr_sf','gr_liv_area']
for col in to_log:
    df['log_' + col] = np.log(df[col])
```

### Get Dummies for Categorical Data

```python
#pd.get_dummies(df.neighborhood).shape 2051, 28
#drop the first dummy
df = pd.get_dummies(df, columns=['neighborhood'], drop_first=True, prefix='dum_nh')
df = pd.get_dummies(df, columns=['ms_zoning'], drop_first=True, prefix='dum_zone')
df = pd.get_dummies(df, columns=['condition_1'], drop_first=True, prefix='dum_cond1')
df = pd.get_dummies(df, columns=['foundation'], drop_first=True, prefix='dum_found')
df = pd.get_dummies(df, columns=['misc_feature'], drop_first=True, prefix='dum_misc')
df = pd.get_dummies(df, columns=['electrical'], drop_first=True, prefix='dum_elect')
df = pd.get_dummies(df, columns=['garage_type'], drop_first=True, prefix='dum_gartype')
```
